# Análisis de las ecuaciones de movimiento

Se obtienen las ecuaciones de movimiento que describen el movimiento del sistema a través de un análisis newtonianiano. Estas ecuaciones son:
 
$m_1L\ddot{\theta_1} = -m_1gsin\left ( \theta_1  \right ) - m_1L \ddot{x}cos\left ( \theta_1 \right ) -\nu_1\dot{\theta_1 }$

$m_2L\ddot{\theta_2} = -m_2gsin\left ( \theta_2  \right ) - m_2L \ddot{x}cos\left ( \theta_2 \right ) -\nu_2\dot{\theta_2 }$

$M\ddot{x}= -\gamma \dot{x}-m_1L\ddot{\theta_1}cos\left ( \theta_1 \right )+m_1L\dot{\theta_1}^2sin\left ( \theta_1 \right )-m_2L\ddot{\theta_2}cos\left ( \theta_2 \right )+m_2L\dot{\theta_2}^2sin\left ( \theta_2 \right )$

Este sistema de ecuaciones diferenciales representa una dificultad para resolverlo de forma analítica en tanto que las ecuaciones estan acopladas y se presentan términos no lineales de las coordenadas $\theta_1, \theta_2$. 
Una primera aproximación, resulta de considerar solo angulos pequeños. Esta suposición es plausible en tanto que de forma experimental se usan ángulos para los que esta aproximación resulta ser buena. En esta aproximación $sin\left ( \theta  \right )\approx \theta $ y $cos\left ( \theta  \right )\approx 1$. 

$m_1L\ddot{\theta_1} = -m_1g \theta_1 - m_1L \ddot{x} -\nu_1\dot{\theta_1 }$

$m_2L\ddot{\theta_2} = -m_2g \theta_2  - m_2L \ddot{x} -\nu_2\dot{\theta_2 }$

$M\ddot{x}= -\gamma \dot{x}-m_1L\ddot{\theta_1}+m_1L\dot{\theta_1}^2 \theta_1 -m_2L\ddot{\theta_2}+m_2L\dot{\theta_2}^2 \theta_2 $


Estas son ecuaciones de segundo orden que se pueden transformar a una forma normal de manera que solo aparezcan ecuaciones de primer orden de forma que sea más sencillo aplicar métodos de solución: 

$m_1L\dot{\phi_1} = -m_1g \theta_1 - m_1L \dot{\chi } -\nu_1\phi_1$

$m_2L\dot{\phi_2} = -m_2g \theta_2  - m_2L \dot{\chi } -\nu_2\phi_2$

$M\dot{\chi}= -\gamma \chi-m_1L\dot{\phi_1}+m_1L\phi_1^2 \theta_1 -m_2L\dot{\phi_2}+m_2L\phi_2^2 \theta_2$ 

$\phi_1 = \dot{\theta_1}$

$\phi_2 = \dot{\theta_2}$

$\chi = \dot{x}$

De estas ecuaciones despenjando las tres primeras para que se pueda expresar de la forma $A\vec{x}+\vec{f} =\dot{ \vec{x}}$ se obtiene:
